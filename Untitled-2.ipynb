```python
import os
import pandas as pd
import sklearn as sk
import sys
import csv
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv("training.csv")
```

```python
data = pd.get_dummies(data, columns = ['wellbore_chev_no_id', 'area_id','formation_id', 'bit_model_id'])
data = data.loc[:, data.columns != 'segment_id']
display(data)
```

```python
train, test = sk.model_selection.train_test_split(data, test_size=0.2)
```

```python
train_cont = train.iloc[:,:6]
train_cont_norm = (train_cont - train_cont.mean())/(train_cont.std())
train_cont_norm.describe()
```

```python
train_norm = pd.concat([train_cont_norm, train.iloc[:,5:]], axis = 1)
```

```python
train_norm.describe()
```

```python
test_cont = test.iloc[:,:6]
test_cont_norm = (test_cont - test_cont.mean())/(test_cont.std())
test_cont_norm.describe()
```

```python
test_norm = pd.concat([test_cont_norm, test.iloc[:,5:]], axis = 1)
test_norm.describe()
```

```python
x_train = train_norm.loc[:, train_norm.columns != 'rate_of_penetration']
x_train = x_train.iloc[:,:6]
x_train.describe()
```

```python
linear = sk.linear_model.LinearRegression()
y_train = train_norm['rate_of_penetration']
```

```python
linear.fit(x_train, y_train)
```

```python
test_x = test_norm.loc[:, test_norm.columns != 'rate_of_penetration']
test_x = test_x.iloc[:,:6]
```

```python
test_y = test_norm['rate_of_penetration']


y_predict = linear.predict(test_x)
print(sk.metrics.r2_score(y_predict, test_y))

print(len(linear.coef_))
plt.plot(y_predict)
```

```python
np.sqrt(sk.metrics.mean_squared_error(test_y, y_predict))
```
